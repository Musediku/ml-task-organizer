```python
# -----------------------------
# Step 1: Import libraries
# -----------------------------
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

# Initialize Faker
fake = Faker()

# -----------------------------
# Step 2: Define parameters
# -----------------------------
NUM_USERS = 10            # number of users
TASKS_PER_USER = 20       # tasks per user
PRIORITY_LEVELS = ['Low', 'Medium', 'High']

STATUS_OPTIONS = ['Completed', 'Pending', 'Not Completed', 'Postponed']
STATUS_PROBS = [0.6, 0.2, 0.1, 0.1]  # probability for each status

# -----------------------------
# Step 3: Generate dataset
# -----------------------------
data = []

for user_id in range(1, NUM_USERS + 1):
    for _ in range(TASKS_PER_USER):
        task_name = fake.sentence(nb_words=5)
        priority = random.choice(PRIORITY_LEVELS)
        status = np.random.choice(STATUS_OPTIONS, p=STATUS_PROBS)
        
        # Random creation date within last 30 days
        created_at = fake.date_time_between(start_date='-30d', end_date='now')
        
        if status == 'Completed':
            # completed tasks
            duration_hours = random.uniform(1, 72)
            completed_at = created_at + timedelta(hours=duration_hours)
        else:
            # Pending, Not Completed, Postponed
            completed_at = pd.NaT
            duration_hours = np.nan
        
        data.append({
            'user_id': user_id,
            'task': task_name,
            'priority': priority,
            'created_at': created_at,
            'completed_at': completed_at,
            'duration_hours': duration_hours,
            'status': status
        })

# -----------------------------
# Step 4: Convert to DataFrame
# -----------------------------
df = pd.DataFrame(data)

# Sort by user and creation date
df = df.sort_values(by=['user_id', 'created_at'])

# -----------------------------
# Step 5: Save to CSV
# -----------------------------
df.to_csv('data/raw/tasks_dataset.csv', index=False)

# -----------------------------
# Step 6: Preview dataset
# -----------------------------
df.head(10)
```
